```python
from text_unidecode import unidecode
from typing import List


def replace_with_similar_latin(text):
    if isinstance(text, str):  # Ensure the input is a string
        return unidecode(text)
    return text


def clean_column(df, column_name):
    """
    Clean the specified column in the DataFrame by converting to lowercase,
    stripping leading/trailing spaces, replacing special characters and spaces,
    and replacing non-Latin characters with similar Latin ones.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the column to clean.
    column_name (str): The name of the column to clean.

    Returns:
    pd.Series: The cleaned column.
    """
    # Apply all cleaning steps
    cleaned_column = (
        df[column_name]
        .apply(replace_with_similar_latin)  # Replace non-Latin characters
        .str.lower()
        .str.strip()
        .str.replace(r"[^\w\s]", "", regex=True)
        .str.replace(" ", "", regex=False)
    )
    return cleaned_column


def clean_review_column(df, column_name):
    """
    Clean the specified review column by replacing various newline characters with spaces
    and normalizing whitespace.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the review column to clean.
    column_name (str): The name of the review column to clean.

    Returns:
    pd.Series: The cleaned review column.
    """
    # Step 1: Ensure the column is of string type. Convert NaN to empty string.
    cleaned = df[column_name].astype(str).fillna("")

    # Step 2: Replace actual newline characters (\n, \r\n, \r) with space
    # This handles Unix, Windows, and old Mac newline characters
    cleaned = cleaned.str.replace(r"(\r\n|\r|\n)", " ", regex=True)

    # Step 3: Replace escaped newline characters (\\n) with space
    # This handles cases where newlines are stored as literal backslash-n
    cleaned = cleaned.str.replace(r"\\n", " ", regex=True)

    # Step 4: Replace multiple consecutive whitespace characters with a single space
    cleaned = cleaned.str.replace(r"\s+", " ", regex=True)

    # Step 5: Strip leading and trailing whitespace
    cleaned = cleaned.str.strip()

    return cleaned


def populate_yandex_addresses(df, address_column):
    """
    Populate the given address column for rows with the same lat/lon using the 'yandex' source as the reference.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - address_column (str): The name of the address column to populate.

    Returns:
    - pd.Series: A pandas Series with updated addresses.
    """
    # Filter rows where the source is 'yandex' and the address is not NaN
    yandex_data = df[df["source"] == "yandex"][["lat", "lon", address_column]]
    yandex_data = yandex_data.dropna(subset=[address_column])

    # Create a mapping of (lat, lon) -> address for yandex rows
    lat_lon_to_address = yandex_data.set_index(["lat", "lon"])[address_column].to_dict()

    # Function to find address based on lat and lon
    def get_address(row):
        key = (row["lat"], row["lon"])
        return lat_lon_to_address.get(
            key, row[address_column]
        )  # Use existing address if no match found

    # Apply the function to populate addresses
    return df.apply(get_address, axis=1)


def populate_target_column(
    df: pd.DataFrame,
    target_column: str,
    grouping_columns: List[str],
    primary_source: str,
    source_column: str = "source",
) -> pd.Series:
    """
    Populate the target column in the DataFrame based on a primary source and grouping columns.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - target_column (str): The name of the column to populate (e.g., "address").
    - grouping_columns (List[str]): Columns to group by (e.g., ["lat", "lon"]).
    - primary_source (str): The source value to prioritize (e.g., "yandex").
    - source_column (str): The name of the source column (default is "source").

    Returns:
    - pd.Series: A pandas Series with updated values for the target column.
    """
    # Step 1: Filter rows where source is primary_source and target_column is not null
    primary_source_data = df[
        (df[source_column] == primary_source) & (~df[target_column].isna())
    ][grouping_columns + [target_column]]

    # Step 2: Drop duplicates to ensure one address per group
    # If multiple entries exist for the same group, the first occurrence is taken
    primary_source_data = primary_source_data.drop_duplicates(subset=grouping_columns)

    # Step 3: Create a mapping from grouping_columns to target_column
    # Convert grouping columns to a tuple to use as a key
    primary_source_data["group_key"] = primary_source_data[grouping_columns].apply(
        tuple, axis=1
    )
    mapping = primary_source_data.set_index("group_key")[target_column].to_dict()

    # Step 4: Create a group key for the entire DataFrame
    df["group_key"] = df[grouping_columns].apply(tuple, axis=1)

    # Step 5: Use the mapping to populate the target column
    # If a mapping exists for the group_key, use it; otherwise, keep the existing value
    df[target_column] = df.apply(
        lambda row: mapping.get(row["group_key"], row[target_column]), axis=1
    )

    # Step 6: Drop the temporary group_key column
    df.drop(columns=["group_key"], inplace=True)

    return df[target_column]


def fillna(
    df: pd.DataFrame, col_name: str, group_cols: List[str], na_type: str
) -> pd.Series:
    """
    Fill NaN values in a specified column using group-wise aggregation.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - col_name (str): The column to fill NaN values for.
    - group_cols (List[str]): Columns to group by.
    - na_type (str): Aggregation type ('mean', 'median', 'min', 'max', etc.).

    Returns:
    - pd.Series: A pandas Series with NaN values filled.
    """
    # Check if na_type is valid
    if na_type not in ["mean", "median", "min", "max", "sum", "std", "var"]:
        raise ValueError(
            f"Invalid na_type: {na_type}. Must be one of: 'mean', 'median', 'min', 'max', 'sum', 'std', 'var'."
        )

    # Group by specified columns and calculate the aggregation
    group_agg = df.groupby(group_cols)[col_name].transform(na_type)

    # Fill NaN values in the column with the calculated group-wise aggregation
    return df[col_name].fillna(group_agg)


def populate_place_id(df: pd.DataFrame, column_name: str) -> pd.Series:
    """
    Replace short place_ids with new unique long IDs in the specified DataFrame column.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the place_ids.
    - column_name (str): The name of the column containing place_ids.

    Returns:
    - pd.Series: The updated place_id column with short IDs replaced by new long IDs.
    """
    # Step 1: Ensure the place_id column is of type string
    df[column_name] = df[column_name].astype(str)

    # Step 2: Identify unique place_ids
    unique_place_ids = df[column_name].unique()

    # Step 3: Separate long and short place_ids
    long_place_ids = [pid for pid in unique_place_ids if pid.isdigit()]
    short_place_ids = [pid for pid in unique_place_ids if not pid.isdigit()]

    # Debug: Print counts
    print(f"Total unique place_ids: {len(unique_place_ids)}")
    print(f"Long place_ids (numeric): {len(long_place_ids)}")
    print(f"Short place_ids (alphanumeric): {len(short_place_ids)}")

    # Step 4: Determine the current maximum long ID
    if long_place_ids:
        max_id = max(int(pid) for pid in long_place_ids)
        print(f"Current maximum long place_id: {max_id}")
    else:
        # Define a starting point if there are no long IDs
        max_id = 80000000000000000  # Adjust as necessary
        print(f"No existing long place_ids found. Starting new IDs from: {max_id}")

    # Step 5: Create a mapping for short place_ids to new long IDs
    new_ids = range(max_id + 1, max_id + 1 + len(short_place_ids))
    mapping = {
        short_pid: str(new_id) for short_pid, new_id in zip(short_place_ids, new_ids)
    }

    # Debug: Show mapping
    print("Mapping of short to new long place_ids:")
    for short_pid, new_long_id in mapping.items():
        print(f"  {short_pid} --> {new_long_id}")

    # Step 6: Replace short place_ids with new long IDs using the mapping
    df[column_name] = df[column_name].replace(mapping)

    # Step 7: Verification
    remaining_short_ids = df[column_name].apply(lambda x: not x.isdigit()).sum()
    if remaining_short_ids > 0:
        print(
            f"Warning: There are still {remaining_short_ids} short place_ids remaining."
        )
    else:
        print("All short place_ids have been successfully replaced with long IDs.")

    return df[column_name]


def calculate_review_count(
    df: pd.DataFrame, place_id_column: str, count_column: str = "reviews_count"
) -> pd.Series:
    """
    Calculate the number of reviews for each place_id and assign it to a specified column.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the place_ids.
    - place_id_column (str): The name of the column containing place_ids.
    - count_column (str): The name of the column to store the review counts. Defaults to 'reviews_count'.

    Returns:
    - pd.Series: The updated review count column.
    """
    # Step 1: Ensure the place_id column is of type string (optional, based on your data)
    df[place_id_column] = df[place_id_column].astype(str)

    # Step 2: Group by place_id and count the number of occurrences
    review_counts = df.groupby(place_id_column)[place_id_column].transform("count")

    # Step 3: Assign the counts to the specified count_column
    df[count_column] = review_counts

    # Optional: Display some information for verification
    unique_places = df[place_id_column].nunique()
    total_reviews = len(df)
    print(f"Calculated review counts for {unique_places} unique place_ids.")
    print(f"Total number of reviews: {total_reviews}")

    return df[count_column]


def truncate_words(df, column):
    def truncate(word):
        if pd.isna(word):
            return word
        word = str(word)
        truncated_length = max(int(len(word) * 0.3), 5)
        return word[:truncated_length]

    return df[column].apply(truncate)
```

```python
test = df_combined.copy()

test["review"] = clean_review_column(test, "review")
test["address"] = populate_target_column(test, "address", ["lat", "lon"], "yandex")
test["name_primary_lower"] = test["name_primary"].str.lower().str.strip()

test["review_lower"] = clean_column(test, "review")
test["name_primary_lower"] = clean_column(test, "name_primary")
test["address_lower"] = clean_column(test, "address")
test["name_primary_trunc"] = truncate_words(test, "name_primary_lower")

test.dropna(subset=["address"], inplace=True)
test.dropna(subset=["name_primary"], inplace=True)

# ---
# populate columns
test["city"] = test["city"].fillna("Москва")

# ---
# populate columns based on filter

# name_extension
test["name_extension"] = populate_target_column(
    test, "name_extension", ["address_lower", "name_primary_lower"], "flamp"
)
test["name_extension"] = populate_target_column(
    test, "name_extension", ["address_lower", "name_primary_lower"], "yandex"
)
test["name_extension"] = populate_target_column(
    test, "name_extension", ["address_lower"], "flamp"
)
test["name_extension"] = populate_target_column(
    test, "name_extension", ["address_lower"], "yandex"
)
test["name_extension"] = test["name_extension"].fillna("ресторан")

# name_primary
test["name_primary"] = populate_target_column(
    test, "name_primary", ["address_lower", "name_primary_lower"], "flamp"
)
test["name_primary"] = populate_target_column(
    test, "name_primary", ["address_lower", "name_primary_lower"], "yandex"
)

# name_primary using trunc method
test["name_primary"] = populate_target_column(
    test, "name_primary", ["address_lower", "name_primary_trunc"], "flamp"
)
test["name_primary"] = populate_target_column(
    test, "name_primary", ["address_lower", "name_primary_trunc"], "yandex"
)

# place_id
test["place_id"] = populate_target_column(
    test, "place_id", ["address_lower", "name_primary_lower"], "flamp"
)
test["place_id"] = populate_target_column(
    test, "place_id", ["address_lower", "name_primary_lower"], "yandex"
)

# ---
# flamp fills yandex missing cols
group_cols = ["address_lower", "name_primary_lower"]

col_name = "business_lunch"
test[col_name] = populate_target_column(test, col_name, group_cols, "flamp")
test[col_name] = test[col_name].fillna(0)

col_name = "avg_price"
test[col_name] = populate_target_column(test, col_name, group_cols, "flamp")
test[col_name] = fillna(test, col_name, group_cols, "mean")

# populate place ids
test["place_id"] = populate_place_id(test, "place_id")

# final cleaning

test["review_lower"] = clean_column(test, "review_lower")
test["name_primary_lower"] = clean_column(test, "name_primary_lower")
test["address_lower"] = clean_column(test, "address_lower")

# display
display_column = "place_id"
search_value = "70000001091774692"
display(
    test[(test[display_column] == search_value) & (test["source"] == "flamp")].head(2)
)
display(
    test[(test[display_column] == search_value) & (test["source"] == "yandex")].head(2)
)
display(test)
display(test.info())
```
